```python
#Import Pakietów
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#Iris data import
from sklearn.datasets import load_iris
irisRaw = load_iris()
```

```python
#Data conversion to pandas format 
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
# 1) Wyświetlenie danych 
iris
```

```python
 # 2) Wyświetl liczbę wierszy i kolumn 
print(f'Liczba wierszy: {iris.shape[0]}')
print(f'Liczba kolumn: {iris.shape[1]}')
```

```python
# 3) Podstawowe informacje dla poszczególnych kolumn 
iris.describe()
```

```python
# 4) Zapoznanie z groupby  
grouped = iris.groupby("target", sort=False) # grouping dataset by target 
grouped.describe()
```

```python
grouped2 = iris.groupby("petal length (cm)") # w tym przypadku użycie groupby nie ma większego sensu 
grouped2.describe()
```

```python
# 5) Wyświetlenie pierwszych 5 wierszy za pomocą metody 'head'
iris.head(5)
```

```python
# 6) Sprawdzenie brakujących danych za pomocą 'dropna' 
if iris.shape == iris.dropna().shape:
  print("Nie ma brakujących danych.")
else:
  print("Brakujące dane pojawiły się.")
```

```python
# 7) Sortowanie danych za pomocą sort_values()
iris.sort_values(iris.columns[1]) # Dane posortowano w zależności od szerokości kielicha
```

```python
# 8) Wyznaczenie minimalnej i maksymalnej długości płatka
print(f"Minimalna długość płatka: {iris['petal length (cm)'].min()}")
print(f"Maksymalna długość płatka: {iris['petal length (cm)'].max()}")
```

```python
# 9) Obliczenie odchylenia standardowego dla każdej kolumny 
[f'Std of {column} is : {round(iris[column].std(), 4)}' for column in iris]
```

```python
# 10)  Wyodrębnij wiersze dla których długość kielicha kwiatów (sepal length) jest większa od średniej długości (cały zbiór)
[index for index, row in iris.iterrows() if row[0] > iris["sepal length (cm)"].mean()] # lista zawiera indexy wierszy w których długość kielicha jest większa od średniej
```

```python
# 11)  Wyświetl histogramy dla poszczególnych parametrów z uwzglednieniem przynależności do danej klasy
colors = ['r', 'b', 'g', 'y', 'skyblue']
for fig_num, column in enumerate(iris):
  plt.figure(figsize=[12, 6], num=fig_num)
  plt.hist(iris[column], color=colors[fig_num], density=True, bins=50)
  plt.title(f'Histogram of {column}')
  plt.xlabel('Value')
  plt.ylabel('Quantity')
  plt.grid()
  plt.show()
```
